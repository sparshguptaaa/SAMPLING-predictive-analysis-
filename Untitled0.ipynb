```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df = X_balanced.copy()
df['Class'] = y_balanced
#Simple Random Sampling
sampling1 = df.sample(frac=0.5, random_state=1)
#Systematic Sampling
k = 2
sampling2 = df.iloc[::k]
#Stratified Sampling
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1)
y = df['Class']

X_s, _, y_s, _ = train_test_split(
    X, y, train_size=0.5, stratify=y, random_state=42
)

sampling3 = X_s.copy()
sampling3['Class'] = y_s
#Cluster Sampling
df['cluster'] = df.index % 5
sampling4 = df[df['cluster'] == 0].drop('cluster', axis=1)
#bootstrap sampling
sampling5 = df.sample(frac=1, replace=True, random_state=42)




```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = sampling1.drop('Class', axis=1)
y = sampling1['Class']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

models = {
    "Model 1_LogisticRegression": LogisticRegression(max_iter=1000),
    "Model 2_DecisionTree": DecisionTreeClassifier(),
    "Model 3_RandomForest": RandomForestClassifier(),
    "Model 4_KNN": KNeighborsClassifier(),
    "Moddel 5_SVM": SVC()
}
samples = {
    "Sampling1": sampling1,
    "Sampling2": sampling2,
    "Sampling3": sampling3,
    "Sampling4": sampling4,
    "Sampling5": sampling5
}

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

performance_table = {}

for sampling_label, data_part in samples.items():

    features = data_part.drop(columns=['Class'])
    target = data_part['Class']

    X_tr, X_te, y_tr, y_te = train_test_split(
        features,
        target,
        test_size=0.3,
        random_state=42
    )

    model_scores = {}

    for model_label, clf in models.items():
        clf.fit(X_tr, y_tr)              # model training
        predictions = clf.predict(X_te) # model prediction
        score = accuracy_score(y_te, predictions)

        model_scores[model_label] = score

    performance_table[sampling_label] = model_scores
```

```python
final_results = pd.DataFrame(performance_table).T
final_results
```
